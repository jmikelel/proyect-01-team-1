# MODELO DE PERCEPTRÓN

### 1. Estructura del Perceptrón
Una red neuronal de una neurona, también conocida como perceptrón, es la forma más simple que puede haber de una red neuronal, ya que está constituida principalmente de una sola neurona que se encarga de hacer todo el trabajo de regresión y clasificación. Su estructura está conformada de las siguientes partes:
- Entrada. Son las variables que corresponden a la entrada de datos que corresponden a la entrada de la neurona.
- Peso. Es la importancia o relevancia que tendrá cada valor de entrada o salida de la neurona. Se suele representar con una ecuación matemática y pueden ser ajustables dependiendo del modelo que se busque realizar
- Sesgo. Corresponde al parámetro adicional que se suma a los valores de pérdida de cada neurona.
- Función de activación. Es la función que se genera de la suma de los parámetros de las entradas, pesos y sesgo previo a la neurona. Las funciones de activación usadas para modelos con perceptrones son: 
- Función de escalón: Produce un resultado binario, es decir, 0 o 1.
   -   Sigmoide: Produce valores entre 0 y 1.
   -   ReLU: Si la entrada es negativa produce 0 y si no lo es entrega el valor en entrada.	
- Salida: Es el resultado que genera la función de activación con la suma de los parámetros.


### 3. Función de Activación
El perceptrón utiliza una función de activación (generalmente la función escalón o la sigmoide) para determinar si una entrada pertenece a una clase. La salida es 1 si la suma ponderada de las entradas supera un umbral; de lo contrario, es 0.

### 4. Cálculo de la Salida
La salida del perceptrón se calcula como:
$$ 
y = f(w_1x_1 + w_2x_2 + ... + w_nx_n + b) 
$$
donde:
- \( y \) es la salida del perceptrón (1 o 0).
- \( w_i \) son los pesos de las entradas.
- \( x_i \) son las entradas.
- \( b \) es el sesgo.
- \( f \) es la función de activación.

### 5. Frontera de Decisión
La **frontera de decisión** es una línea (en 2D) o un hiperplano (en dimensiones superiores) que separa las dos clases. En un espacio de dos dimensiones, se puede visualizar como:

- **Ecuación de la frontera de decisión**:
$$
w_1x_1 + w_2x_2 + b = 0
$$

Los puntos a un lado de la línea (por ejemplo, arriba) serán clasificados como una clase (1), mientras que los puntos al otro lado (abajo) serán clasificados como otra clase (0).

### 6. Entrenamiento del Perceptrón
Durante el entrenamiento, el perceptrón ajusta sus pesos y el sesgo a través de un proceso iterativo llamado **algoritmo de retropropagación** o **regla de aprendizaje del perceptrón**. Esto se realiza ajustando los pesos según la regla:

$$
w_i \leftarrow w_i + \eta (y_{\text{real}} - y_{\text{predicho}}) x_i
$$

donde:
- \( \eta \) es la tasa de aprendizaje.
- \( y_{\text{real}} \) es la etiqueta real (1 o 0).
- \( y_{\text{predicho}} \) es la salida del perceptrón.

### 7. Ejemplo de Clasificación
Imaginemos que tenemos un conjunto de datos con dos características (por ejemplo, largo y ancho de pétalos) y queremos clasificar flores en dos tipos (por ejemplo, "setosa" y "versicolor").

1. **Inicialización**: El perceptrón comienza con pesos y sesgo aleatorios.
2. **Entrenamiento**: Ajusta los pesos basándose en ejemplos del conjunto de datos, buscando la mejor línea (frontera de decisión) que separe las dos clases.
3. **Clasificación**: Una vez entrenado, clasifica nuevas entradas observando de qué lado de la frontera de decisión caen.

### 8. Visualización
En una representación gráfica, si los datos de entrenamiento están bien separados, la frontera de decisión se verá claramente, dividiendo las dos categorías. Sin embargo, si los datos no son linealmente separables, el perceptrón simple no podrá encontrar una frontera de decisión adecuada, y podría requerirse un modelo más complejo, como un perceptrón multicapa.

## FUNCIONES DE ACTIVACIÓN
Las funciones de activación son esenciales para las redes neuronales porque introducen no linealidad en el modelo, lo que permite a las redes aprender patrones complejos. Se utilizan generalmente varios tipos de funciones de activación, dependiendo de si el problema es de clasificación (asignación de una categoría a una entrada) o de regresión (predicción de valores continuos). A continuación, discutiremos las funciones de activación para cada tipo de tarea mas comunes.

### Funciones de Activación para Clasificación
Las funciones de activación transforman los valores de las neuronas en problemas de clasificación para que puedan representar probabilidades o decisiones de clase

1. **Sigmoide (Logística):**
   - **Fórmula:**
   $$\sigma(x) = \frac{1}{1 + e^{-x}}$$
   - **Descripción:** Para problemas de clasificación binaria (dos clases), la función sigmoide convierte los valores de entrada en una salida que oscila entre 0 y 1. Se utiliza con frecuencia en la capa superior de las redes neuronales para clasificación binaria porque representa una probabilidad.
   - **Problemas:** Tiene el problema del desvanecimiento del gradiente, que retrasa el entrenamiento de redes neuronales profundas. Este problema surge cuando los gradientes se reducen significativamente a valores extremos.
   - **Aplicación:** Clasificación binaria (problema de 2 clases).
   

2. **Softmax:**
   - **Fórmula:**
   $$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}$$
   - **Descripción:** Es una función que transforma un conjunto de salidas en probabilidades para que todas sumen 1. Cuando se trata de problemas de clasificación multiclase, es útil. Softmax normaliza las activaciones de la capa de salida para que cada valor muestre la probabilidad de pertenecer a una clase específica.
   - **Aplicación:** Clasificación multiclase, es decir, clasificación de más de dos clases.


3. **Tanh (Tangente Hiperbólica):**
   - **Fórmula:**
   $$\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$
   - **Descripción:** A diferencia de la Sigmoide, la tangente hiperbólica escala las salidas entre -1 y 1, en lugar de entre 0 y 1. Esto puede mejorar la convergencia durante el entrenamiento al hacer que las activaciones sean más “centradas” alrededor de cero.
   - **Problemas:** Aunque es una alternativa superior a la sigmoide para redes profundas, tiene problemas con el desvanecimiento del gradiente.
   - **Aplicación:** Clasificación binaria y multiclase, aunque menos común que ReLu o Softmax.



### Funciones de Activación para Regresión
En problemas de regresión, las salidas del modelo son valores continuos, y se suelen usar funciones de activación que no limiten tanto los valores de salida.

1. **ReLU (Rectified Linear Unit):**
   - **Fórmula:**
   $$\text{ReLU}(x) = \max(0, x)$$
   - **Descripción:** ReLU es una de las funciones de activación más utilizadas porque es simple y efectiva. Al pasar los valores negativos a 0, introduce una no linealidad y deja pasar los valores positivos como están. No satura como la Sigmoide o el Tanh, lo que reduce el problema del desvanecimiento del gradiente.
   - **Problemas:** Si muchos valores negativos se mapean a 0 y los pesos no se actualizan, puede tener el problema de “neurona muerta”.
   - **Aplicación:** En problemas de regresión y clasificación son comunes para redes neuronales profundas. Muchas redes neuronales la tienen por defecto.

2. **Leaky ReLU:**
   - **Fórmula:**
$$\text{LeakyReLU}(x) = 
\begin{cases} 
x & \text{si } x \geq 0 \\
\alpha x & \text{si } x < 0
\end{cases}$$
   - **Descripción:** Leaky ReLU es una variante de ReLU en la que, para evitar el problema de neuronas muertas, se le asigna un pequeño valor negativo (α) en lugar de convertirlos a 0. Por ejemplo,  puede ser 0.01.
   - **Aplicación:** Útil para problemas de regresión y clasificación, especialmente en redes profundas donde ReLU puede causar problemas de neuronas muertas.

3. **Linear (Lineal):**
   - **Fórmula:**
   $$\text{Linear}(x) = x$$
   - **Descripción:** Esta función trivial deja los datos tal como están en lugar de transformarlos. Es apropiado para problemas de regresión en los que se espera que las salidas sean valores continuos en lugar de clases o probabilidades. Como no introduce no linealidad, se utiliza con frecuencia en la capa de salida de los modelos de regresión.
   - **Aplicación:** Regresión (predicción de valores continuos)

```python
# Set a global seed for reproducibility
tf.random.set_seed(4500) # Se fija una semilla para obtener respuestas similares al trabajar con números aleatorios.
```

```python
X = df.drop(["target"], axis=1) # Se elimina la columna occupancy para las entradas (X).
y = df['target']  # El occupancy se convierte en la salida o Y, para predecir si esta ocupada o no la habitación.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True) # Se separan los datos de entrenamiento y prueba, corresponiendo el 80% de los datos a el entrenamiento y el resto para prueba, con un random state de 42, para obtener respuestas similares al trabajar con aleatoriedad, además de la inclusión de la función shuffle, ayuda a que exista más esparcimiento entre los datos.
```

Se optó por una división de 70/30 entre los datos de entrenamiento y de prueba, donde se obtuvieron mejores resultados que con una división de 80/20.

```python
# Standardise features
scaler = StandardScaler() # Herramienta para estandarizar datos y evitar sesgos.
X_train = scaler.fit_transform(X_train) # Estandariza los datos de x train.
X_test = scaler.transform(X_test) # # Estandariza los datos de x test.
```

```python
# Define the model

def get_model():
    model = tf.keras.Sequential([  # Se crea un modelo secuencial de Keras.
        tf.keras.layers.InputLayer(shape=(len(X.columns),)), # Se establecen 5 entradas para: temperatura, humedad, luz, co2 y relación de humedad.
        tf.keras.layers.Dense(1, activation='sigmoid' ) # Capa con solamente una neurona, y con la función de activación sigmoidal.
    ])
    return model

model = get_model() # Se obtiene el modelo.
```

Para este proyecto, se hicieron pruebas con las funciones de activación sigmoidal y tanh, donde se obtuvieron mejores resultados con la función sigmoidal. Esto sucede debido a que la función sigmoidal obtiene valores de salida entre 0 y 1. Por el otro lado, la función tanh, obtiene valores de salida entre -1 a 1, donde para una clasificación binaria, podríamos definir y < 0.5 = 0 y y > 0.5 = 1, mientras que la función tanh maneja una mayor cantidad de números.

```python
model.summary() # Se devuelve un análisis del modelo.
```

Se observa un resumen del modelo de nuestra red neuronal de una sola neurona, la cual cuenta con 14 parámetros, donde 13 son las entradas y 1 es la salida, los cuales ocuparán 56 Bytes de memoria, más del doble de los utilizados en la actividad anterior.

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test) # Se evalua la precisión del modelo con el conjunto de entrenamiento antes de entrenar.
print(f"Test Accuracy: {accuracy:.4f}") # Se muestran los resultados.
```

Para la compilación del modelo se estableció el optimizador Adam, el cual combina lo mejor de los algoritmos Momentum y RMSProp, este tiene una implementación sencilla y pocos requisitos de memoria, lo que la convierte en una opción preferible en la mayoría de situaciones, es el más utilizado en la industria. Para el learning rate, se hicieron pruebas con 3 learning rates distintos, (0.001, 0.003 y 0.01), donde se obtuvieron mejores resultados con un alpha = 0.001, para aumentar el learning rate, se aplicó la metodología del profesor Andrew Ng, donde este sugiere que se aumente multiplicándolo o diviendo este entre 3. Asimismo, se estableció la función de pérdida crossentropy, la cual es utilizada principalmente para clasificaciones binarias.

```python
from keras.callbacks import EarlyStopping

# Definir el callback de Early Stopping
early_stopping = EarlyStopping(
    patience=50,
    monitor='val_loss',
    min_delta=0.001,
    restore_best_weights=True,
)

from tensorflow.keras.callbacks import LearningRateScheduler

# Función para duplicar el learning rate una sola vez después de la época 70
def scheduler(epoch, lr):
    if epoch == 70:
        return lr / 1.5
    return lr  # Mantiene el mismo learning rate después de la duplicación

# Definir el callback de LearningRateScheduler
lr_scheduler = LearningRateScheduler(scheduler)

# Entrenar el modelo con el callback que minimiza el learning rate
history = model.fit(
    X_train, y_train,
    epochs=600,
    batch_size=32,
    validation_split=0.20,
    callbacks=[early_stopping, lr_scheduler]  # Añadir el callback de EarlyStopping scheduler
)
```

Para el entrenamiento se optó por dejar el batch_size en 32, cabe destacar que se hicieron pruebas con 16 y 32, obteniendo los mejores resultados con un batch de 32. Además, se aplicó el callback EarlyStopping para reducir la cantidad de épocas necesarias para encontrar los parámetros deseados. Asimismo, en la clase vimos como podemos ajustar la pérdida, en este caso, a partir de aproximadamente la época 70 (no lo fue en esta iteración) en diversas pruebas, la pérdida comenzaba a aumentar, con lo visto en clase, esto se puede explicar debido a que como el valor de alpha es muy grande, el modelo en el descenso del gradiente empieza a dar pasos muy elevados, por lo que, para intentar que este llegué a una convergencia, se optó por reducir un poco más el learning rate alpha/1.5, se dividió entre 1.5 debido al mismo criterio propuesto por el profesor Andrew Ng, donde en este caso en lugar de dividirlo entre 3, se divido por la mitad de este, para que no sea un cambio tan brusco. El número de épocas se determinó en este caso de manera aleatoria, algo que la función EarlyStopping se encarga de ajustar.

```python
# Plot validation and testing accuracy
plt.figure(figsize=(12,6)) # Se establece el tamaño de la gráfica.
plt.plot(history.history['accuracy']) # Se gráfica la exactitud del modelo a través de las épocas.
plt.plot(history.history['val_accuracy']) # Se gráfica la exactitud del modelo con los datos de validación del entrenamiento a través de las épocas.
plt.xlabel('epoch') # El eje x corresponde a las épocas.
plt.ylabel('accuracy') # El eje y corresponde a las exactitud.
plt.legend(['train', 'val']) # Se añade una leyenda para distinguir entre entrenamiento y validación.
plt.grid(); # Se añade una cuadrícula a la gráfica para mejorar la legibilidad.
```

```python
# Plot validation and testing loss
plt.figure(figsize=(12,6)) # Se establece el tamaño de la gráfica.
plt.plot(history.history['loss']) # Se gráfica la pérdida del conjunto de entrenamiento a lo largo de las épocas.
plt.plot(history.history['val_loss']) # Se gráfica la pérdida con los datos de validación del entrenamiento a lo largo de las épocas.
plt.xlabel('epoch') # El eje x corresponde a las épocas.
plt.ylabel('loss') # El eje y corresponde a la pérdida.
plt.legend(['train', 'val'])  # Se añade una leyenda para distinguir entre entrenamiento y validación.
plt.grid(); # Se añade una cuadrícula a la gráfica para mejorar la legibilidad.
```

En la matriz de confusión se observa cómo existe una mayor cantidad de falsos negativos que falsos positivos, resultados los cuales no se buscan en este tipo de aplicaciones, donde es preferible un diagnóstico erróneo a correr el riesgo de generar una falsa alarma. Es preferible alertar a un paciente sano que comprometer la detección de una condición crítica.

```python
def get_performance_metrics(c_matrix):
    tn, fp, fn, tp = c_matrix[0][0], c_matrix[0][1], c_matrix[1][0], c_matrix[1][1] # Se obtienen los datos de la matriz de confusión, (verdaderos negativos, falsos positivos, falsos negativos y verdaderos positivos).

    accuracy = (tp + tn) / (tp + tn + fp + fn) # Exactitud.
    precision = tp / (tp + fp) # Precisión.
    recall = tp / (tp + fn) # Sensibilidad o Recall.
    specificity = tn / (tn + fp) # Especificidad.
    f1_score = 2 * (precision * recall)/(precision + recall) # F1-score.
    return accuracy, precision, recall, specificity, f1_score # Se devuelven los valores de las métricas de rendimiento.

accuracy, precision, recall, specificity, f1_score = get_performance_metrics(c_matrix) # Se obtienen las métricas de desempeño usando la matriz de confusión.

# Se imprimen los resultados con 6 decimales
print(f"Accuracy: {accuracy:0.6f}")
print(f"Precision: {precision:0.6f}")
print(f"Recall: {recall:0.6f}")
print(f"Specificity: {specificity:0.6f}")
print(f"F1-score: {f1_score:0.6f}")
```
